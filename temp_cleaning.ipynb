```python
def cleaner1(index, drop):
    dic = {}
    n = 0
    dic[0] = []
    exc = ['', '배달팁', '(A) 매출금액', '(B) 고객할인비용', '(C) 차감금액', '(D) 부가세', '(E) 만나서결제금액', '입금예정금액 (A+B+C+D)', '입금예정금액 (A+B+C+D+E)']
    exc.append(drop)

    for i in data[index][1]:
        if i == '총 결제금액':
            n += 1
            dic[n] = []
            pass
        elif i in exc:
            pass
        else:
            dic[n].append(i)
    space = len(dic)-1
    del dic[space]
    return list(dic.values())
```

## 

```python
def dfmaker(index):
    df = pd.concat((pd.DataFrame([i.split('\n') for i in data[index][0]]), pd.DataFrame(pd.Series(cleaner1(index, drop='┗ 기본')))), axis=1)
    df.columns = ['date', 'time', 'menu']
    return df
```

```python
df_total = dfmaker(0)
df_total = df_total[df_total['date']=='2022. 06. 23. (목)']
df_total
```
