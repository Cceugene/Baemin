# Pickle로 데이터 불러오기

```python
import pickle

with open('./time.pkl', 'rb') as f:
    times = pickle.load(f)

times
```

# 가게별로 나누기

```python
date_NM = times[:104]
date_KB = times[104:184]
date_CH = times[184:217]
date_TK = times[217:]
```

```python
len(date_TK)
```

# 날짜와 시간으로 나누기

```python
import pandas as pd
```

## 가게 전체

```python
date_list = [date_NM, date_KB, date_CH, date_TK]
NM = []
KB = []
CH = []
TK = []
vari = [NM, KB, CH, TK]
```

```python
for i, v in zip(date_list, vari):
    for t in i:
        e = t.split('\n')
        v.append(e)
```

```python
print(len(vari[3]))
vari[3]
```

## 달리는 타코야끼_지난 달

```python
with open('time_tako.pkl', 'rb') as f:
    dates_TAKO = pickle.load(f)
```

```python
date_TAKO = []

for i in dates_TAKO:
    a, b = i.split('\n')
    box = [a[:13], a[15], b[:2], b[3:11]]
    date_TAKO.append(box) 
```

```python
date_TAKO
```

```python
df = pd.DataFrame(date_TAKO, columns=['Date', 'Week', 'Noon', 'Time'])
```
