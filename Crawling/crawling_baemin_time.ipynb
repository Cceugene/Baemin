```python
from selenium import webdriver
import time
```

```python
chrome_driver = './chromedriver.exe'
driver = webdriver.Chrome(chrome_driver)
login_url = 'https://ceo.baemin.com/web/login?returnUrl=https%3A%2F%2Fceo.baemin.com%2F'
url = 'https://ceo.baemin.com/self-service/orders/history'
driver.get(login_url)
time.sleep(1)

# Login
ID = driver.find_element_by_name('id')
PASS = driver.find_element_by_name('password')
 # id, password
ID.send_keys('')
PASS.send_keys('')
button = driver.find_element_by_xpath('//*[@class="Button__StyledButton-sc-394bsp-0 iqPqWd"]')
button.click()
time.sleep(1)

# 페이지 이동
driver.get(url)
time.sleep(2)

# 기간 선택
driver.find_element_by_xpath('//*[@class="FilterContainer-module__dOxp FilterContainer-module__dam2"]').click()
time.sleep(1)
driver.find_element_by_xpath('//div[@class="Radio-module__P8H0"]/label[2]/p').click() # 월 별 기간
driver.find_element_by_css_selector('.DefaultDateFilterDialog-module__zF8H div div div:nth-child(2) label input').click() # 지난 달
driver.find_element_by_xpath('//button[@class="button medium"]').click()

# 가게 선택
stores = ['13260990', '13643240', '13442351', '13708011'] # 냉면, 국밥, 치밥, 타코야끼
dates = []
for i in stores:
    time.sleep(1)
    driver.find_element_by_xpath('//*[@class="FilterContainer-module__ZmLA"]').click()
    time.sleep(1)
    driver.find_element_by_xpath(f'//select/option[@value="{i}"]').click()
    driver.find_element_by_xpath('//button[@class="button medium"]').click()

    for r in range(21):
        try:
            time.sleep(1)
            
            # scrapping (날짜/시간)
            ds = driver.find_elements_by_css_selector('table tr td:nth-child(2)')

            # 리스트에 넣기
            for d in ds:
                dates.append(d.text)

            # 다음 페이지
            time.sleep(2)
            button_next = driver.find_element_by_xpath('//a[@class="Paging-module__BwVC"]/i[@class="icon icon-arrow-right"]')
            button_next.click()
        except:
            break
    print(len(dates))
    
print('완료')
```
