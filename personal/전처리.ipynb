## 데이터 전처리

```python
## 함수 정의

# 임시 전화번호 결측처리 함수
def nanmaker(x):
    if x[:4] == '0503':
        out = np.nan
    else:
        out = x
    return out

# 데이터프레임 만드는 함수
def dfmaker(data):
    df = data[~data['도착지'].isnull()][['도착지', '음식요금', '배달요금', '거리(km)', '요청시간', '픽', '진행시간', '도착지 전화번호', '기타']]
    df['도착지 전화번호'] = df['도착지 전화번호'].apply(nanmaker)
    return df
```

```python
df1 = dfmaker(data1)
df2 = dfmaker(data2)
df3 = dfmaker(data3)
df4 = dfmaker(data4)
df5 = dfmaker(data5)
df6 = dfmaker(data6)
df7 = dfmaker(data7)
df7
```

```python
df = pd.concat([df1, df2, df3, df4, df5, df6, df7])
df = df.drop_duplicates()
df.sort_values('요청시간').reset_index().drop('index', axis=1)
```
